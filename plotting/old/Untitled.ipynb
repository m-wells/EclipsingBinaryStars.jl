```julia
using Optim
```

```julia
δ = 1.0
q = 0.5
ε = 0.0
#F = EclipsingBinaryStars.get_syncpar(ε)
F = 1
pots = [ 6.0
       , 3.5
       , 2.87584
       , 2.8
       , 2.7
       ]
plotcolors = [ "blue"
             , "green"
             , "black"
             , "magenta"
             , "red"
             ]
xL1 = EclipsingBinaryStars.get_xL1(0.5,1.0)
```

```julia
function Ω(ϱ,θ)
    λ = sin(θ)
    ν = cos(θ)
    EclipsingBinaryStars.get_Ω(ϱ,δ,q,λ,ν,F)
end
```

```julia
Ω(xL1,π/2)
```

```julia
pots = [ 6.0 ,3.5
       , Ω(xL1,π/2)
       , 2.5 , 2.0
       ]
```

```julia
n = 1000
θs = collect(linspace(0, 2π,n))
#θs = collect(linspace(0,π,n))
ϱs = Array{Float64,1}(n)
;
```

```julia
function get_ϱlim(xL1, ϕ)
    # ϕ measured from CW from +z
    if ϕ < π/4
        return 1.0
    elseif ϕ < 3π/4
        return 1.2*xL1/cos(π/2 - ϕ)
    else
        return 1.0
    end
end
```

```julia
for (plotcolor,pot) in zip(plotcolors,pots)
    for (i,θ) in enumerate(θs)
        f(x) = abs(pot - Ω(x,θ))
        
        ϱlim = get_ϱlim(xL1,θ)
        
        res = optimize(f, 0, ϱlim, Brent())
        #println(Optim.converged(res))
        if Optim.converged(res)
            ϱs[i] = Optim.minimizer(res)
        else
            ϱs[i] = NaN
        end
    end
    
    xs = @. ϱs*sin(θs)
    zs = @. ϱs*cos(θs)
    
    mask = xs .< xL1
    
    xs = xs[mask]
    zs = zs[mask]
    
    plot( xs
        , zs
        , color = plotcolor
        , marker = "o"
        , markersize = 0.5
        , linestyle = "None"
        )
end
```

```julia
for (plotcolor,pot) in zip(plotcolors,pots)
    for (i,θ) in enumerate(θs)
        potprim = pot/q + 0.5*(q-1)/q
        f(x) = abs(pot - Ω(x,θ))
        #f(x) = abs(potprim - Ω(x+1,θ))
        
        ϱlim = get_ϱlim(xL1,θ)
        
        res = optimize(f, ϱlim, 5, Brent())
        #println(Optim.converged(res))
        if Optim.converged(res)
            ϱs[i] = Optim.minimizer(res)
        else
            ϱs[i] = NaN
        end
    end
    
    xs = @. ϱs*sin(θs) + 1
    zs = @. ϱs*cos(θs)
    
    mask = xs .> xL1
    
    xs = xs[mask]
    zs = zs[mask]
    
    plot( xs
        , zs
        , color = plotcolor
        , marker = "o"
        , markersize = 0.5
        , linestyle = "None"
        )
end
```
